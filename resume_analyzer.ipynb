```python
# Resume Analyzer (NLP)
Analyze a resume, extract skills, match with job roles, and suggest skill gaps.
```

```python
# Import Libraries
import re
import spacy

nlp = spacy.load("en_core_web_sm")
```

```python
# Sample Resume Text
text = """
I have experience in Python, Machine Learning, Data Analysis, SQL, and Deep Learning.
Worked on NLP projects and data visualization using Power BI.
"""
```

```python
# Text Cleaning
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    return text

cleaned_text = clean_text(text)
print("Cleaned Text:\n", cleaned_text)
```

```python
# Tokenization
doc = nlp(cleaned_text)

tokens = [token.text for token in doc]
print("\nTokens:\n", tokens)
```

```python
# Skill List
skills_list = [
    "python", "machine learning", "data analysis", "sql",
    "deep learning", "nlp", "power bi", "excel", "tableau"
]
```

```python
# Skill Extraction
def extract_skills(text, skills_list):
    found = []
    for skill in skills_list:
        if skill in text:
            found.append(skill)
    return found

found_skills = extract_skills(cleaned_text, skills_list)
print("\nExtracted Skills:\n", found_skills)
```

```python
# Job Role Database
job_roles = {
    "Data Scientist": ["python", "machine learning", "deep learning", "nlp"],
    "Data Analyst": ["sql", "excel", "power bi", "data analysis"],
    "ML Engineer": ["python", "machine learning", "deep learning"]
}
```

```python
# Job Matching and Skill Gap Analysis
def analyze_jobs(found_skills, job_roles):
    results = {}

    for job, skills in job_roles.items():
        matched = set(found_skills) & set(skills)
        missing = set(skills) - set(found_skills)
        score = len(matched) / len(skills)

        results[job] = {
            "matched": matched,
            "missing": missing,
            "score": score
        }

    return results

results = analyze_jobs(found_skills, job_roles)
```

```python
# Display Results
for job, info in sorted(results.items(), key=lambda x: x[1]["score"], reverse=True):
    print("\n==============================")
    print("Job Role:", job)
    print("Match Score:", round(info["score"] * 100, 2), "%")
    print("Matched Skills:", info["matched"])
    print("Missing Skills:", info["missing"])
```

```python
# Best Job Recommendation
best_job = max(results.items(), key=lambda x: x[1]["score"])

print("\n==============================")
print("Best Suitable Job Role:", best_job[0])
```
